```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def normal_distribution(x, mean, std_dev):
    # Calculate the probability density function (PDF) of the normal distribution
    pdf = norm.pdf(x, mean, std_dev)
    return pdf

def plot_normal_curve(mean, std_dev, num_points=100):
    # Generate x values for plotting the curve
    x = np.linspace(mean - 3 * std_dev, mean + 3 * std_dev, num_points)
    print(x)
    # Calculate the corresponding probability density function (PDF) values
    pdf = normal_distribution(x, mean, std_dev)

    # Plot the normal curve
    plt.figure(figsize=(8, 6))
    plt.plot(x, pdf, label=f"μ={mean}, σ={std_dev}")
    plt.title('Normal Distribution Curve')
    plt.xlabel('x')
    plt.ylabel('Probability Density Function (PDF)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
mean = 0  # Mean of the normal distribution
std_dev = 1  # Standard deviation of the normal distribution
plot_normal_curve(mean, std_dev)
```
